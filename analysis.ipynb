```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('../Sales-Data-Analysis-SQL/sample_sales_data.csv')

# Basic EDA
print(df.head())
print(df.describe())

# Visualization - Revenue by Category
sns.barplot(x='Category', y='Revenue', data=df)
plt.title('Revenue by Category')
plt.show()

# Regression Model - Predict Revenue based on Quantity & Price
X = df[['Quantity', 'Price']]
y = df['Revenue']
model = LinearRegression()
model.fit(X, y)
print('Regression Coefficients:', model.coef_)
print('Intercept:', model.intercept_)

# Probability Distribution of Revenue
sns.histplot(df['Revenue'], kde=True)
plt.title('Revenue Distribution')
plt.show()
```
